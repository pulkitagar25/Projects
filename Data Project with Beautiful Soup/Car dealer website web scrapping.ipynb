### Importing Libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib.parse 
```

### HTTP Request

```python
website="https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D=&list_price_max=&maximum_distance=20&zip="
```

```python
response= requests.get(website)
```

```python
response.status_code
```

### Soup Object

```python
soup=BeautifulSoup(response.content,'html.parser')
```

### Results

```python
results = soup.find_all('div', {'class':'vehicle-card'})
```

```python
len(results)
```

### Target necessary data

```python
#Product Name
results[0].find('h2').get_text()
```

```python
#Mileage
results[0].find('div',{'class':'mileage'}).get_text()
```

```python
#Primary price
results[0].find('span',{'class':'primary-price'}).get_text()
```

```python
#Discount price
results[0].find('span',{'class':'secondary-price price-drop'}).get_text()
```

```python
#Product Rating
results[0].find('span',{'class':'sds-rating__count'}).get_text()
```

```python
#Product Review
results[0].find('span',{'class':'sds-rating__link'}).get_text()
```

```python
#Dealer-Name
results[0].find('div',{'class':'dealer-name'}).get_text().strip()
```

### Running a for loop 

```python
product_name=[]
product_mileage=[]
product_price=[]
product_dis_price=[]
product_rating=[]
product_review=[]
product_dealer=[]

for result in results:
    #product_name
    try:
        product_name.append(result.find('h2').get_text())
    except:
        product_name.append('n/a')
        
    #product_mileage
    try:
        product_mileage.append(result.find('div',{'class':'mileage'}).get_text())
    except:
        product_mileage.append('n/a')
        
    #product_price
    try:
        product_price.append(result.find('span',{'class':'primary-price'}).get_text())
    except:
        product_price.append('n/a')
        
    #product_dis_price
    try:
        product_dis_price.append(result.find('span',{'class':'secondary-price price-drop'}).get_text())
    except:
        product_dis_price.append('n/a')
        
    #product_rating
    try:
        product_rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
    except:
        product_rating.append('n/a')
        
    #product_review
    try:
        product_review.append(result.find('span',{'class':'sds-rating__link'}).get_text())
    except:
        product_review.append('n/a')
        
    #product_dealer
    try:
        product_dealer.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
    except:
        product_dealer.append('n/a')
```

### Create Pandas Dataframe

```python
car_product = pd.DataFrame({'Name': product_name,'Mileage':product_mileage, 'Price':product_price,'Discount':product_dis_price, 'Rating':product_rating,
                                'Review Count': product_review, 'Dealer': product_dealer})
```

```python
car_product
```

#### Data Cleaning

```python
car_product['Review Count'] = car_product['Review Count'].apply(lambda x: x.strip('reviews)').strip('('))
car_product['Discount'] = car_product['Discount'].apply(lambda x: x.strip('price drop)').strip('('))
```

## For Multiple page

```python
product_name=[]
product_mileage=[]
product_price=[]
product_dis_price=[]
product_rating=[]
product_review=[]
product_dealer=[]


for i in range (1,20):
    
    website = 'https://www.cars.com/shopping/results/?page='+ str(i) +'&page_size=100&dealer_id=&list_price_max=&list_price_min=&makes[]=mercedes_benz&maximum_distance=20&mileage_max=&sort=best_match_desc&stock_type=cpo&year_max=&year_min=&zip=' 
    
    response = requests.get(website)
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    results = soup.find_all('div', {'class' : 'vehicle-card'})

    for result in results:
    #product_name
        try:
            product_name.append(result.find('h2').get_text())
        except:
            product_name.append('n/a')
        
    #product_mileage
        try:
            product_mileage.append(result.find('div',{'class':'mileage'}).get_text())
        except:
            product_mileage.append('n/a')
        
    #product_price
        try:
            product_price.append(result.find('span',{'class':'primary-price'}).get_text())
        except:
            product_price.append('n/a')
        
    #product_dis_price
        try:
            product_dis_price.append(result.find('span',{'class':'secondary-price price-drop'}).get_text())
        except:
            product_dis_price.append('n/a')
        
    #product_rating
        try:
            product_rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
        except:
            product_rating.append('n/a')
            
    #product_review
        try:
            product_review.append(result.find('span',{'class':'sds-rating__link'}).get_text())
        except:
            product_review.append('n/a')
        
    #product_dealer
        try:
            product_dealer.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
        except:
            product_dealer.append('n/a')
```

```python
## dictionary
car_product = pd.DataFrame({'Name': product_name,'Mileage':product_mileage, 'Price':product_price,'Discount':product_dis_price, 'Rating':product_rating,
                                'Review Count': product_review, 'Dealer': product_dealer})
```

### Data cleaning

```python
car_product['Review Count'] = car_product['Review Count'].apply(lambda x: x.strip('reviews)').strip('('))
car_product['Discount'] = car_product['Discount'].apply(lambda x: x.strip('price drop)').strip('('))
```
